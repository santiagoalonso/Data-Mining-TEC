# Hypotheses as distributions

```python
age_sample = np.random.normal(loc = 34, scale = 10, size = 1000)
age_hypothesis_A = np.random.normal(loc = 20, scale = 5, size = 1000)
age_hypothesis_B = np.random.normal(loc = 40, scale = 20, size = 1000)

sns.kdeplot(age_sample, label = 'sample')
sns.kdeplot(age_hypothesis_A, label = 'HA')
sns.kdeplot(age_hypothesis_B, label = 'HB')
plt.title("Which hypothesis would you believe?")
plt.legend();
```

```python
Null_hypothesis = np.random.standard_t(20, 10000)

sns.kdeplot(Null_hypothesis, label = 'Null hypothesis t(20)')
plt.vlines(x = 3, ymin = 0, ymax = 0.3, label = 'Data', color = 'red')
plt.legend()
plt.title('The standard t distribution');
```

Before anything, lets practice how to load and clean data. Also, some exploratory data analysis (EDA)


```python
#Load data
#This first data (https://www.kaggle.com/datasets/amirmotefaker/ab-testing-dataset)
# is separated intro control (AB_control_group.csv) and test group (AB_test_group.csv).
#               Campaign Name: The name of the campaign
#               Date: Date of the record
#               Spend: Amount spent on the campaign in dollars
#               of Impressions: Number of impressions the ad crossed through the campaign
#               Reach: The number of unique impressions received in the ad
#               # of Website Clicks: Number of website clicks received through the ads
#               # of Searches: Number of users who performed searches on the website
#               # of View Content: Number of users who viewed content and products on the website
#               # of Add to Cart: Number of users who added products to the cart
#               # of Purchase: Number of purchases
# load both datasets. Name them AB_website_test and AB_website_control, respectively
# tip (in case you need it): in pd.read_csv(), you can change the delimitation character of columns with sep.
AB_website_test = pd.read_csv("AB_test_group.csv", sep = ";")
AB_website_control = pd.read_csv("AB_control_group.csv", sep = ";")

# Change the column names in both as follows:
# Campaign Name -> Campaign
# Date -> Date
# Spend [USD] -> Cost_Campaign
# # of Impressions -> n_Impresions
# Reach -> Reach
# # of Website Clicks -> n_Clicks
# # of Searches -> n_Searches
# # of View Content -> n_Views
# # of Add to Cart -> n_Cart
# # of Purchase -> n_Purchase
cols = ["Campaign", "Date", "Cost_Campaign", "n_Impresions", "Reach", 
        "n_Clicks", "n_Searches", "n_Views", "n_Cart", "n_Purchase"]
AB_website_test.columns = cols
AB_website_control.columns = cols

#Merge both data sets with pd.concat() 
#Call it AB_website. Remember to .reset_index(drop=True)
AB_website = pd.concat([AB_website_test, AB_website_control]).reset_index(drop=True)

#Print the data AB_website 
#Which column has the treatment information.
print(AB_website.head())

#Print data types of AB_website
print(AB_website.dtypes)


#Plot a chart with the distributions of n_Cart.
#Each campaign separately, but on the same plot 
#Use seaborn, .kdeplot or .histplot 
#Fill the distributions (or bars) with a color for each campaign and change the alpha level i.e. transparency
#Any preliminary reflections regarding the efficacy of the campaign?
sns.kdeplot(data=AB_website, x = "n_Cart", hue = "Campaign", fill = "Campaign", alpha = 0.5)
#sns.histplot(data=AB_website, x = "n_Cart", hue = "Campaign", fill = "Campaign", alpha = 0.2)


#Now start another figure with plt.figure()
#Plot the same plot as before but now for n_Purchase. 
#Any new reflections regarding the efficacy of the campaign?
plt.figure()
sns.kdeplot(data=AB_website, x = "n_Purchase", hue = "Campaign", fill = "Campaign", alpha = 0.5)

#Lets see all the variables now.
#Use .groupby() to calculate the mean, std, min, and max of all numerical columns
#tip: look for .agg()
#Group by the column "Campaign"
#Put the grouping in a variable called AB_descriptive
funs = ["mean", "std",'min', 'max']
AB_descriptive = AB_website.groupby(["Campaign"]).agg({'Cost_Campaign': funs,
                                                       'n_Impresions': funs,
                                                       'Reach': funs,
                                                       'n_Clicks': funs,
                                                       'n_Searches': funs,
                                                       'n_Views': funs,
                                                       'n_Cart': funs,
                                                       'n_Purchase': funs})
#Print from AB_descriptive the results of n_Clicks and n_Purchase
#Tip: use .loc
#Any new preliminary reflections?
print(AB_descriptive.loc[:, ["n_Clicks", "n_Purchase"]])

#What data type is the column Date? 
#Create a new column in AB_website. Call it Date_Formatted. Put it next to Date. Tip: use .insert()
#The content of Date_Formatted should be the equal to column Date but in pandas date format. 
#Tip: pd.to_datetime(). The format is '%d.%m.%Y'
#Create a new column in AB_website. Call it Date_Day with just the day of Date_Formatted
#tip: use .dt.day on Date_Formatted
AB_website.insert(2, "Date_Formatted", 
                  pd.to_datetime(AB_website.loc[:, "Date"], format='%d.%m.%Y'))
AB_website.insert(2, "Date_Day", 
                  AB_website.loc[:, "Date_Formatted"].dt.day)


#Plot n_Cart (y) across days (x). Each campaign a different colored line, but in the same plot.
#Tip: use seaborn
#Change the labels of the axis to something more expressive
#Comment on the dynamics? Are both groups similar? Explain
plt.figure()
sns.lineplot(AB_website, x = "Date_Day", y = "n_Cart", hue = "Campaign");
plt.xlabel("Time (Days)")
plt.ylabel("Users who added a product to the cart")


```

This image ([source: Brian OGorman](https://github.com/brian-c-ogorman/ABanditTesting/)) has four groups during testing. The best group ends up being 1 and is selected/exploited henceforth. Thus, all the remaining groups experienced suboptimal options (regret) during the test/exploratory phase.

<img src="A-B_Test.png" width="600"/>

For instance, the 4 options could be different movies. The platform, say Netflix, wants to determine what to recommend to some users with shared characteristics. Netflix then does A/B testing, meaning, user pool 1 watches movie 1, user pool 2 watches movie 2, and so forth. Asumming perfect comparability between the 4 user pools, Netflix finds that movie 1 got higher scores (in binary, more 5 out of 5 because for sure they liked it). How do we know, statistically, that the proportion of 5 out of 5 is indeed larger than the other groups?
